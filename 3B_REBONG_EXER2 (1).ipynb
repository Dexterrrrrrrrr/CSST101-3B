# Assessment Task 1: Basic Probability Calculations


```python
def joint_probability(p_a, p_b):

    return p_a * p_b

def marginal_probability(p_a, p_b, joint_p_ab):


    return p_a + (p_b - joint_p_ab)

def conditional_probability(joint_p_ab, p_b):

    if p_b == 0:
        raise ValueError("P(B) cannot be zero for conditional probability calculation.")

    return joint_p_ab / p_b
```

```python
p_A = 0.3
p_B = 0.4

joint_p_AB = joint_probability(p_A, p_B)
print(f"Joint Probability P(A ∩ B) when A and B are independent: {joint_p_AB:.4f}")
```

# Assessment Task 2: Bayesian Inference


**Interactive Bayesian Model:**

```python
def bayesian_inference(prior_prob, likelihood_pos, likelihood_neg, evidence_pos):

    if evidence_pos:
        likelihood = likelihood_pos
    else:
        likelihood = 1 - likelihood_pos

    posterior_prob = (prior_prob * likelihood) / (prior_prob * likelihood + (1 - prior_prob) * (1 - likelihood))

    return posterior_prob
```

**Interactive Scenario:**

```python
prior_prob = 0.01
likelihood_pos = 0.90
likelihood_neg = 0.05

posterior_prob = bayesian_inference(prior_prob, likelihood_pos, likelihood_neg, True)

print(f"Probability of having the disease given a positive test result: {posterior_prob:.4f}")
```

# Assessment Task 3: Decision-Making Under Uncertainty

```python
import random

def investment_simulation(iterations, probability_success, reward, penalty):

    total_return = 0

    for _ in range(iterations):
        # Simulate the investment decision
        if random.random() < probability_success:
            total_return += reward
        else:
            total_return += penalty

    average_return = total_return / iterations

    return average_return
```

```python
average_return = investment_simulation(1000, 0.7, 1000, -500)
print(f"Average return after 1000 iterations: ${average_return:.2f}")
```

# Assessment Task 4: Real-World Scenario

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def predict_rain(humidity, cloud_cover):
    humidity = humidity / 100
    cloud_cover = cloud_cover / 100

    if humidity > 0.5 and cloud_cover > 0.5:
        return 0.8
    elif humidity > 0.5 and cloud_cover <= 0.5:
        return 0.6
    elif humidity <= 0.5 and cloud_cover > 0.5:
        return 0.7
    else:
        return 0.2

def simulate_rain_prediction():
    humidities = np.linspace(0, 100, num=100)
    cloud_covers = np.linspace(0, 100, num=100)

    predictions = np.zeros((len(humidities), len(cloud_covers)))

    for i, humidity in enumerate(humidities):
        for j, cloud_cover in enumerate(cloud_covers):
            predictions[i][j] = predict_rain(humidity, cloud_cover)

    # Create contour plot
    plt.figure(figsize=(12, 6))

    # Contour plot
    plt.subplot(1, 2, 1)
    plt.contourf(cloud_covers, humidities, predictions.T, levels=np.linspace(0, 1, 11), cmap='Blues')
    plt.colorbar(label='Probability of Rain')
    plt.title('Probability of Rain Based on Humidity and Cloud Cover')
    plt.xlabel('Cloud Cover (%)')
    plt.ylabel('Humidity (%)')
    plt.xlim(0, 100)
    plt.ylim(0, 100)
    plt.grid(True)

    # Scatter plot for specific points
    selected_humidities = [20, 40, 60, 80]
    selected_cloud_covers = [10, 30, 50, 70]

    scatter_predictions = [predict_rain(h, c) for h, c in zip(selected_humidities, selected_cloud_covers)]

    plt.subplot(1, 2, 2)
    plt.scatter(selected_cloud_covers, selected_humidities, c=scatter_predictions,
                cmap='Reds', s=100, edgecolor='black', label='Predicted Rain Probability')

    # Add colorbar for scatter plot
    scatter = plt.scatter([], [], c=[], cmap='Reds', s=100)
    cbar = plt.colorbar(scatter)
    cbar.set_label('Probability of Rain')

    plt.title('Selected Humidity and Cloud Cover')
    plt.xlabel('Cloud Cover (%)')
    plt.ylabel('Humidity (%)')

    for i in range(len(selected_humidities)):
        plt.annotate(f'{scatter_predictions[i]:.2f}',
                     (selected_cloud_covers[i], selected_humidities[i]),
                     textcoords="offset points",
                     xytext=(5,-5),
                     ha='center')

    plt.xlim(0, 100)
    plt.ylim(0, 100)

    plt.grid(True)

    # Show plots
    plt.tight_layout()
    plt.show()

simulate_rain_prediction()
```

#  Assessment Task 5: Probability Distributions Visualization

**Interactive Visualization Task:**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, norm

def plot_binomial(n, p):
    x = np.arange(0, n+1)
    pmf = binom.pmf(x, n, p)

    plt.figure(figsize=(12, 6))
    plt.bar(x, pmf, color='skyblue', alpha=0.7, label=f'Binomial Distribution (n={n}, p={p})')
    plt.title('Binomial Probability Distribution')
    plt.xlabel('Number of Successes')
    plt.ylabel('Probability')
    plt.xticks(x)
    plt.legend()
    plt.grid()
    plt.show()

def plot_normal(mu, sigma):
    x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
    pdf = norm.pdf(x, mu, sigma)

    plt.figure(figsize=(12, 6))
    plt.plot(x, pdf, color='orange', label=f'Normal Distribution (μ={mu}, σ={sigma})')
    plt.title('Normal Probability Distribution')
    plt.xlabel('Value')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid()
    plt.show()

plot_binomial(n=10, p=0.5)
plot_normal(mu=0, sigma=1)
```

```python
def simulate_coin_flips(n_trials, p_heads):
    flips = np.random.binomial(n_trials, p_heads)


    outcomes = np.arange(0, n_trials + 1)
    probabilities = [np.random.binomial(n_trials, p_heads) for _ in range(1000)]

    plt.figure(figsize=(12, 6))
    plt.hist(probabilities, bins=n_trials+1, color='lightgreen', alpha=0.7)
    plt.title(f'Simulation of {n_trials} Coin Flips (p={p_heads})')
    plt.xlabel('Number of Heads')
    plt.ylabel('Frequency')
    plt.xticks(outcomes)
    plt.grid()
    plt.show()

simulate_coin_flips(n_trials=10, p_heads=0.5)
```
